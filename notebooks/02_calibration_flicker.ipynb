# 02 — Breakpoint flicker and the bp/ decision (CLAUDE.md §7.3)

The blocked-EM approximation (§3.5) holds an edge's contribution constant over its
span; the residual is **breakpoint flicker** at persist-but-reparent boundaries.
Decision criterion: build loopy BP/EP (`bp/`) only if flicker is large relative to
the true-tract discontinuity. We split each query's segment-boundary flicker into
boundaries that coincide with a true ancestry switch vs. those that do not.

```python
import numpy as np
from tslai.experiments import admixture_experiment, flicker_vs_true_boundaries

r = admixture_experiment(T_admix=300, n_admix=6, n_ref=10, sequence_length=4e5,
                         f_A=0.5, max_iter=8, seed=1)
off, at = [], []
for q in r['tracks']:
    fv = flicker_vs_true_boundaries(r['tracks'], r['truth_states'], q)
    off.append(fv['mean_flicker_off_true'])
    if not np.isnan(fv['mean_flicker_at_true']):
        at.append(fv['mean_flicker_at_true'])
print(f'flicker at NON-true boundaries (persist-but-reparent): {np.mean(off):.4f}')
print(f'discontinuity at TRUE switches:                       {np.mean(at):.4f}')
print(f'ratio: {np.mean(off) / np.mean(at):.4f}')
```

## Decision

On these (favorable, true-ARG) sims, flicker at non-true boundaries is ~0.001 while
the discontinuity at true switches is ~0.95 — three orders of magnitude smaller.
By the §7.3 criterion **blocked EM is sufficient and `bp/` is not needed**; it stays
a footnote. Re-run this notebook on inferred ARGs (Relate/tsinfer) and weak-structure
or ancient-admixture scenarios — the regimes where the approximation is most stressed —
before treating the decision as final.
